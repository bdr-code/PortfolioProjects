```python
# First let's import the packages we will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the CSV Data
df = pd.read_csv(r'C:\Users\dell\Desktop\movies.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df = df.dropna()
```

```python
# data types for our column

df.dtypes
```

```python
# We can change data types of colums

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
df = df.sort_values(by =['gross'], inplace=False, ascending=False)
```

```python
# We can drop any Duplicates in the dataframe

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
# We can also Drop any duplicates accross the entire dataframe
df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y =df['gross'])
plt.title('Budget vs Gross Earning')

plt.xlabel('Budget for film')
plt.ylabel('Gross Revenue')

plt.show()
```

```python
df.head()
```

```python
# budget vs gross using seaborn
sns.regplot(x ='budget', y ='gross', data = df, scatter_kws ={'color':'red'}, line_kws = {'color': 'blue'})
```

```python
df.corr(method = 'pearson') #pearson, kendall, spearman
```

```python
# 'budget' and 'gross' has a correlation of 0.74 which is the highest.

#'gross' has the Second highest Correlation with 'votes' at 0.61
```

```python
correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix , annot = True)
plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df.head()
```

```python
df_numerized =  df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name]  = df_numerized[col_name].cat.codes
df_numerized.head()                                           
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)> 0.5]
high_corr
```
